```python
import requests
from bs4 import BeautifulSoup
import os
import re

# SITEMAP_URL = "https://www.davidkolbconsultancy.com/pages-sitemap.xml"

SITEMAP_URL = "https://www.davidkolbconsultancy.com/pages-sitemap.xml"

HEADERS = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/89.0.4389.82 Safari/537.36"
}
OUTPUT_DIR = "downloaded_pages"

# Ensure output directory exists
if not os.path.exists(OUTPUT_DIR):
    os.makedirs(OUTPUT_DIR)

# Fetch the sitemap content
sitemap_response = requests.get(SITEMAP_URL, headers=HEADERS)
sitemap_soup = BeautifulSoup(sitemap_response.content, 'xml')

# Extract URLs from the sitemap
urls = [loc.text for loc in sitemap_soup.find_all("loc")]

# Choose a specific URL to test (e.g., the first URL in the list)
test_url = urls[0]

# Fetch content of the chosen URL and save to a separate text file
response = requests.get(test_url, headers=HEADERS)
page_soup = BeautifulSoup(response.content, 'html.parser')

# Extract the title from the HTML
title_tag = page_soup.find("title")
if title_tag:
    title = title_tag.get_text(strip=True)
else:
    title = "Untitled"  # Use a default title if no <title> tag is found

# Additional text to be added at the end of the file
additional_text = '"'

# Extract text content from the HTML and remove consecutive blank lines
text_content = page_soup.get_text(separator='\n').strip()
text_content = re.sub(r'\n{2,}', '\n\n', text_content)


# Create a filename based on the URL
filename = test_url.rstrip('/').split('/')[-1] or "index"
filepath = os.path.join(OUTPUT_DIR, f"{filename}.txt")

with open(filepath, 'w', encoding="utf-8") as file:
    # Write the title and additional text to the file
    file.write("Page Title : " + '"' + title)
    file.write(text_content + "\n")
    file.write(additional_text)

print(f"Saved content of {test_url} to {filepath}")
```
